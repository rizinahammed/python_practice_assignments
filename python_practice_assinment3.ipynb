1. Write a Python program which accepts the radius of a circle from the user

3. Write a Python program to get the largest number from a list.


```python
numbers = [12, 3, 47, 10]
largest = numbers[0]

for n in numbers:
    if n > largest:
        largest = n
print(largest, "is the largest number.")
```

4.  Given a range of first 10 numbers, write a Python program to iterate from
start number to the end number and print the sum of the current number
and previous number.  

```python
previous = 0
for current in range(1, 11):
    total = current + previous
    print("Current Number", current, "Previous Number", previous, "Sum:", total)
    previous = current
```

5. Write a Python program to print only those numbers which are divisible
of 5.  

```python
numbers = [10, 20, 33, 46, 55]

for n in numbers:
    if n % 5 == 0:
        print(n)
```

6. Write a Python program to check whether a number is prime or not.


```python
num = int(input("Enter the number: "))

if num <= 1:
    print(num, "is not a prime number.")
else:
    is_prime = True
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num, "is a prime number.")
    else:
        print(num, "is not a prime number.")
```

7. Write a Python program to reverse a list using for loop.  

```python
numbers = [10, 40, 30, 70]
reversed_list = []

for i in range(len(numbers)-1, -1, -1):
    reversed_list.append(numbers[i])
print(reversed_list)
```

nested for loop.

```python
for i in range(1, 6):
    for j in range(i):
        print("*", end="")
    print()
for i in range(4, 0, -1):
    for j in range(i):
        print("*", end="")
    print()
```
